```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## random

```python
def create_random_data(n_components=5, amplitude_range=(0.1, 1.0), frequency_range=(5, 20), noise_level=1, seed=42):
    if seed is not None:
        np.random.seed(seed)

    time = np.linspace(0, 1, 100, endpoint=False)
    data = np.zeros(len(time))
    for _ in range(n_components):
        amplitude = np.random.uniform(*amplitude_range)
        frequency = np.random.uniform(*frequency_range)
        phase = np.random.uniform(0, 2 * np.pi)

        data += amplitude * np.sin(2 * np.pi * frequency * time + phase)
    data += np.random.normal(loc=0, scale=noise_level, size=len(data))
    return data
```

```python
data = create_random_data(
    n_components=5,
    amplitude_range=(0.1, 1.0),
    frequency_range=(5, 20),
    noise_level=.5,
    seed=42,
)

plt.figure(figsize=(6, 4))
# plt.plot(data, c='k', label='Random Frequency Components + Noise')
plt.plot(data, label='Random Frequency Components + Noise')
plt.title('Time Series Data with Random Frequency Components')
plt.xlim(0, len(data))
plt.ylim(-3.5, 3.5)
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```

## sin

```python
def create_random_sin_data(n_components=5, amplitude_range=(0.1, 1.0), frequency_range=(5, 20), seed=42):
    if seed is not None:
        np.random.seed(seed)

    time = np.linspace(0, 1, 100, endpoint=False)
    data = np.zeros(len(time))
    for _ in range(n_components):
        amplitude = np.random.uniform(*amplitude_range)
        frequency = np.random.uniform(*frequency_range)
        phase = np.random.uniform(0, 2 * np.pi)

        data += amplitude * np.sin(2 * np.pi * frequency * time + phase)
    return data
```

```python
data = create_random_sin_data(
    n_components=5,
    amplitude_range=(0.1, 1.0),
    frequency_range=(5, 20),
    seed=42,
)

plt.figure(figsize=(6, 4))
plt.plot(data, c='k', label='Random Frequency Components')
# plt.title('Time Series Data with Random Frequency Components')
plt.xlim(0, len(data))
plt.ylim(-3.5, 3.5)
plt.gca().xaxis.set_visible(False)  # X軸を非表示にする
plt.gca().yaxis.set_visible(False)  # Y軸を非表示にする
# plt.grid()
plt.tight_layout()
# plt.legend()
plt.show()
```
